```python
class Node:
  def __init__(self, my_data):
    self.data = my_data
    self.next = None
```

```python
class LinkedList:
  def __init__(self):
    self.head = None
```

```python

lList = LinkedList()
```

```python
i = 0
while i < 5 :
 my_data = int(input("Enter Number: "))
 New_Node = Node(my_data)
 New_Node.next = lList.head
 lList.head = New_Node
 i = i + 1
```

```python
my_link = lList.head
while (my_link != None) :
  print(my_link.data)
  my_link = my_link.next
```
