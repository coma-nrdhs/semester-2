List Definition

```python
class Node:
  def __init__(self, my_data):
    self.data = my_data
    self.next = None
```

```python
class Linked_List:
  def __init__(self):
    self.head = None
```

```python
def insert_front(my_list, new_data) :
    new_node =Node(my_data)
    new_node.next = my_list.head
    my_list.head = new_node
```

```python
 def print_list(my_list) :
    my_link = my_list.head
    while (my_link != None) :
      print(my_link.data)
      my_link = my_link.next
```

```python
def search_list(my_list, value):
   my_link = my_list.head
   i = 0
   while (my_link != None) :
      if my_link.data == value:
        return i
      my_link = my_link.next
      i = i + 1
   else :
      return -1
```
